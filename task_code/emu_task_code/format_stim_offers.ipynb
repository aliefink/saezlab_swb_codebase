```python
new_loss = current_loss.copy().drop_duplicates(subset='Bet 1')
new_loss = new_loss[(new_loss['Bet 1']%0.03 == 0) | (new_loss['Bet 1']%0.5 == 0) | (new_loss['Bet 1'] < -1.45)].drop_duplicates(subset='Bet 1')
new_loss = new_loss.reset_index(drop=True)
# new_loss['Stim_Trial'] = np.arange(current_loss['Trial'].iloc[0],current_loss['Trial'].iloc[0]+len(new_loss))
new_loss['Stim_Trial'] = np.arange(new_gain['Stim_Trial'].iloc[-1]+1,new_gain['Stim_Trial'].iloc[-1]+1+len(new_loss))
new_loss
```
